# Data process

```python
import pandas as pd
df = pd.read_csv("/Users/damo/Documents/Study/DATA5002/Assignment 2/data/movies_metadata.csv")
df.head()
```

## Miss value

```python
df = df[df["vote_count"] >= 100]
df = df[df["runtime"] > 0]
df = df[df["revenue"] > 0]
df[df["genres"] != "[]"]
df
```

## Format transformation

### Genres 

```python
import ast
df['genres_list'] = df['genres'].apply(lambda x: [d['name'] for d in ast.literal_eval(x)])
df_long = df.explode('genres_list')
df_wide = df_long.assign(val=1).pivot_table(
    index='id',
    columns='genres_list',
    values='val',
    fill_value=0
).reset_index()

df_wide.columns = ['id'] + ['genres_' + str(col) for col in df_wide.columns if col != 'id']

df = pd.merge(df,df_wide,on="id")
df
```

```python
df_wide_col = [col for col in df_wide.columns if col.startswith("genres_")]
df_wide_col
```

### Country

```python
import ast
df['Countries_list'] = df["production_countries"].apply(lambda x: [d['name'] for d in ast.literal_eval(x)])
df_long = df.explode('Countries_list')
df_wide = df_long.assign(val=1).pivot_table(
    index='id',
    columns='Countries_list',
    values='val',
    fill_value=0
).reset_index()

df_wide.columns = ['id'] + ['Countries_' + str(col) for col in df_wide.columns if col != 'id']

df = pd.merge(df,df_wide,on="id")
df
```

```python
df_wide_col = [col for col in df_wide.columns if col.startswith("Countries_")]
df_wide_col
```

### Spoken language

```python
import ast
df['Spoken_languages_list'] = df["spoken_languages"].apply(lambda x: [d['name'] for d in ast.literal_eval(x)])
df_long = df.explode('Spoken_languages_list')
df_wide = df_long.assign(val=1).pivot_table(
    index='id',
    columns='Spoken_languages_list',
    values='val',
    fill_value=0
).reset_index()

df_wide.columns = ['id'] + ['Spoken_languages_' + str(col) for col in df_wide.columns if col != 'id']

df = pd.merge(df,df_wide,on="id")
df
```

```python
df_wide_col = [col for col in df_wide.columns if col.startswith("Spoken_languages_")]
df_wide_col
```

# Data Export
